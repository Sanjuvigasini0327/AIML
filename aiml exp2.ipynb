```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_excel("AIML Exp.1.xlsx")
df
```

```python
df = df.drop("S.No", axis=1)
df.head()
```

```python
df = df.drop("category", axis=1)
df.head()
```

```python
plt.xlabel("POWER RATING")
plt.ylabel("AMOUNT OF COPPER")
plt.title("COPPER PREDICTION IN GENERATOR")
plt.hist(df[['power_rating', 'weight_of_copper']])
```

```python
from sklearn.model_selection import train_test_split

x = np.array(df['power_rating'])
y = np.array(df['weight_of_copper'])

x = x.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
len(x_train)
```

```python
len(y_train)
len(x_test)
len(y_test)  
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
x_pred = model.predict(x_test)
model.score(x_test, y_test)
model.predict([[6.29]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

x = np.array(df['power_rating'])
y = np.array(df['weight_of_copper'])

x = x.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x)

plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.plot(x, y_pred)

plt.xlabel("POWER RATING")
plt.ylabel("WEIGHT OF COPPER")
plt.title("LINEAR REGRESSION: COPPER PREDICTION")

plt.show()
```

```python
from sklearn.metrics import r2_score

y_test_pred = model.predict(x_test)
r2 = r2_score(y_test, y_test_pred)

print("R2 Score:", r2)
```
